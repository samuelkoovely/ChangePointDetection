```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import auxiliary_functions
```

```python
selected_lamdas = np.logspace(-5,0,10)
```

```python
with open('block1activity_sparse.pkl', 'rb') as handle:
    tnet = pickle.load(handle)[0]
```

```python
net = tnet['tnet']
```

```python
net.compute_laplacian_matrices(save_adjacencies=True, random_walk=False)
number_active_events = []
for i,t in enumerate(net.times[:-1]):
    number_active_events.append(np.sum(net.adjacencies[i].toarray()) / 2)
```

```python
window = 1
Conditional_S_selected1 = []

for i, lamda in enumerate(selected_lamdas):
    with open(f'/scratch/tmp/180/skoove/block1activity_sparse/net{0}/window_S/{window}/window_S{lamda:.11f}', 'rb') as f:
        S_rate = pickle.load(f)
        Conditional_S_selected1.append(S_rate['signal'])
t_sampled1 = (S_rate['t_samples'])
```

```python
####### Enotrpy Plot
fig, ax = plt.subplots()
color = 'tab:red'
ax.plot(net.times[:-1], number_active_events, color=color)

ax.set_xlabel("t [s]")
ax.set_ylabel("Number of links", color=color)
ax.tick_params(axis='y', labelcolor=color)



ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis

index = 9
S = Conditional_S_selected1[index]

color = 'tab:blue'
ax2.plot(t_sampled1, S, color=color, alpha=1)
ax2.set_xlabel("t [s]")
ax2.set_ylabel("Entropy", color=color)
ax2.tick_params(axis='y', labelcolor=color)


plt.tight_layout()  # otherwise the right y-label is slightly clipped
#plt.savefig('/home/b/skoove/Desktop/evolving_SBM/evolving_SBM_2activities.png', format='png', dpi=300, bbox_inches='tight')
```
